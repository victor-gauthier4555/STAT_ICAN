```python
import pandas as pd
import random as rd
from scipy.stats import normaltest
import warnings
```

```python
data  = pd.read_csv("Exemple_ligne.csv", sep = ";")
data
```

```python
a = (data.iloc[:, 2]).dtype
a
```

```python
index = []
i = 1
a = None
while a != 'float64'  and i < len(data.columns) :
    a = (data.iloc[:, i]).dtype

    if a == 'O' :
        print("OK")

        index.append(i)
    i = i + 1
index
```

```python
random_indices = rd.sample([i for i in range(max(index), len(data.columns))], k = 10)
random_indices
```

```python
indexes = [i for i in range(max(index) + 1 , len(data.columns))] ; indexes
```

```python
random_indices = [3, 8, 10, 4, 11, 5, 13, 17, 9, 6]

for indexo in index :
    print(f'Nous en somme à l index {indexo}')
    L = []
    for indice in random_indices :
        
        groups = data.groupby(data.columns[indexo]).groups
        n = 0
        for key in groups :
            d = data.loc[groups[key], data.columns[indice]]  # This selects column 6 from the grouped data
            # Suppressing the warning
            with warnings.catch_warnings():
                warnings.simplefilter("ignore")
                a = normaltest(d)
            
                if a[1] < 0.05 :
                    n = n + 1
            L.append(n)
    print(L)
    
```
