## Writing a CSV file with the pathways corresponding to the significant metabolites

```python
import pandas as pd
import os 
from math import isnan
```

```python
pathway = pd.read_csv("PATHWAYS.csv", sep = ';',low_memory = False,  header=None, encoding = 'utf-8' ) ; pathway
```

```python
data = pd.read_csv("SIGNIFICANTS_1.csv", sep = ';') ; data
```

```python
with open("test.csv", "w") as file :
    file.write("Pathway name;Coverage (%);Total Metabolites;Retrieved Metabolites\n")
    for k in range(0, len(data.columns))  :
        L = []
        for i in range(0, len(pathway[0])) :
            n = 0
            N = []
            ligne = pathway.iloc[i]
            ligne = ligne[ligne.notna() == True ]
            for metabolite in data.iloc[:,k] : 
                if pd.isna(metabolite) == False : 
                    for j in range(1,len(ligne)) :
                        if pd.isna(ligne[j]) == False :
                            if metabolite == ligne[j]:
                                N.append(metabolite)
                                n += 1
                                if ligne[0] not in L :
                                    L.append(ligne[0])

            if n > 0 :
                N = ','.join(N)
                file.write(f"{ligne[0]};{n/(len(ligne) - 1) * 100};{len(ligne) - 1};{N}\n")

        
                                
                    
    
        
```
